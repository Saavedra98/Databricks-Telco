#### Crear External Locations requeridas para el proyecto
1. Bronze
2. Silver
3. Gold

```sql
CREATE EXTERNAL LOCATION IF NOT EXISTS sadatabricksextsaer_bronze
    URL 'abfss://bronze@sadatabricksextsaer.dfs.core.windows.net/'
    WITH (STORAGE CREDENTIAL `databricks-ext-storage-credential`)
```
