#### CREAR TABLAS EN SCHEMA BRONZE

```sql
DROP TABLE IF EXISTS telco_dev.bronze.clientes;
```

```sql
DROP TABLE IF EXISTS telco_dev.bronze.clientes;

CREATE TABLE IF NOT EXISTS telco_dev.bronze.clientes (
  clienteId INT,
  nombre STRING,
  paterno STRING,
  materno STRING,
  documentoIdentidad STRING,
  telefono STRING,
  email STRING,
  direccion STRING
)
USING CSV
OPTIONS(path "abfss://bronze@sadatabricksextsaer.dfs.core.windows.net/clientes.csv", delimiter ";", header "true")
```

```sql
SELECT * FROM telco_dev.bronze.clientes;
```

```sql
DROP TABLE IF EXISTS telco_dev.bronze.detalles;

CREATE TABLE IF NOT EXISTS telco_dev.bronze.detalles (
  detalleId INT,
  clienteId INT,
  servicio STRING,
  plan STRING,
  precioMensual DECIMAL(5,2),
  estado BOOLEAN,
  fechaInicio DATE,
  fechaFin DATE
)
USING CSV
OPTIONS(path "abfss://bronze@sadatabricksextsaer.dfs.core.windows.net/detalles.csv", delimiter ";", header "true")
```

```sql
SELECT * FROM telco_dev.bronze.detalles;
```
